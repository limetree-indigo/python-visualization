# 히트맵 예제
- 타이타닉호의 연령대_객실등급별 승선자수

## 데이터 준비

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

## 데이터 전처리

```python
# 결측치 확인
titanic.isnull().sum()
```

```python
# 컬럼 삭제(결측치가 많은 열 삭제)
titanic = titanic.drop(columns=['deck'])
```

```python
# 결측치 삭제(결측치가 있는 행 삭제)
titanic = titanic.dropna()
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

### 연령대 컬럼 생성

```python
titanic
```

```python
titanic_cp = titanic.copy()
```

```python
titanic_cp['agerange'] = (titanic_cp['age']/10).astype('int')*10
```

```python
titanic_cp
```

### 피벗테이블 : 연령대-객실등급 별 승선자수

```python
titanic_pivot = titanic_cp.pivot_table(index='class',columns='agerange', values='survived', aggfunc='count')
```

```python
titanic_pivot
```

```python
# 모든 NaN 을 0으로 변경
titanic_pivot = titanic_pivot.fillna(0)
titanic_pivot
```

```python
# 모든 자료의 타입을 int로 변경
titanic_pivot = titanic_pivot.astype('int')
titanic_pivot
```

## 히트맵
### matplotlib

```python
range(0,len(titaic_pivot.columns),1)
```

```python
np.arange(0.5,len(titaic_pivot.columns),1)
```

```python
plt.pcolor(titaic_pivot)
plt.colorbar()

plt.xticks(np.arange(0.5,len(titaic_pivot.columns),1), labels=titaic_pivot.columns)
plt.yticks(np.arange(0.5,len(titaic_pivot.index),1), labels=titaic_pivot.index)

plt.show()
```

### seaborn
- heatmap(data=2차원데이터)<br>
    cmap=컬러맵: 컬러맵 지정<br>
    annot=True: 수치 표시<br>
    fmt='d': 정수로 표시

```python
sns.heatmap(titanic_pivot, cmap='Blues', annot=True, fmt='d')
```
