# 스캐터플롯
- 점의 크기, 색깔로 정보를 표현할 수 있다.(버블차트)
- 크기는 s, 색깔은 c로 지정한다.

```python
# 데이터
x = [1,5,6,9,10]
y = [1,5,3,9,7]
```

## 기본

```python
plt.scatter(x, y)
```

```python
plt.plot(x,y,'o')
```

## 점의 크기
**`s=크기공통지정`**<br>
**`s=크기목록`**
- plt.scatter()의 파라미터 s로 점 크기를 지정한다.
- 점 크기는 공통된 크기로 지정할 수도 있고, 점마다 다른 크기로 지정할 수도 있다.

### 공통 크기 지정

```python
plt.scatter(x,y,s=100)
```

### 값에 따른 다른 크기 지정

```python
plt.scatter(x,y,s=[100,200,50,80,30])
```

## 점의 색상
**`c=색상공통지정`**<br>
**`c=색상목록`**
- plt.scatter()의 파라미터 c로 점 색상을 지정한다.
- 점 크기는 공통된 색상으로 지정할 수도 있고, 점마다 다른 색상을 지정할 수도 있다.

### 단일색 지정

```python
plt.scatter(x,y,s=[100,200,500,80,30], c='r')
```

### 값에 따른 다른 색상 지정
- ['r','b','g','k','y']

```python
plt.scatter(x,y,s=[100,200,500,80,30], c=['r','b','g','k','y'])
```

### 컬러맵으로 색상 지정
- 컬러맵 지정 : cmap=컬러맵
- 컬러바 표시: plt.colorbar()
- 컬러맵 : https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html

```python
plt.scatter(x,y,c=y,cmap='cool')
plt.colorbar()
```

```python
# 컬러맵 종류
plt.colormaps()
```

## 지불금액과 팁의 관계
- 지불금액이 클수록 팁도 많이 줄까?

## 점의 크기로 테이블 인원 표시

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*50,alpha=0.5)
```

## 점의 색으로 성별 표시
