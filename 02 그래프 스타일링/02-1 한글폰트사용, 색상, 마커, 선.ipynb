# 한글폰트 사용

```python
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'AppleGothic' in f.name]
```

```python
# 한글폰트 지정
plt.rcParams['font.family'] = 'AppleGothic'

# 한글폰트 사용시 -기호 깨지는 문제해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 그래프로 확인
x = ['봄', '여름','가을','겨울']
y = [19,30,15,-10]

# 그래프 그리기
plt.plot(x,y)
```

# 색상, 마커, 선

## 색상
- color=색상
- 색상 이름을 사용한다. https://matplotlib.org/3.5.2/gallery/color/named_colors.html
- 자주 사용되는 색깔은 약자를 사용할 수 있다.(blue:b, green:g, red:r, cyan:c, magenta:m, black:k, white:w)
- hex code를 사용할 수 있다.

```python
x = ['봄', '여름', '가을','겨울']
y = [19,30,15,-10]
```

```python
# 색상이름
plt.plot(x,y, color='skyblue')
```

```python
# 색상이름 약자(blue:b, green:g, red:r, cyan:c, magenta:m, black:k, white:w)
plt.plot(x,y,color='r')
```

```python
# 색상코드
plt.plot(x,y,color='#222')
```

## 마커
- maker: 마커종류(*.,o v^ < > 1 2 3 4 s p * h H + x D d)
- makersize, ms: 마커사이즈
- makeredgecolor, mec: 마커 선 색깔
- makeredgewidth, mew: 마커 선 굵기
- makerefacecolor, mfc: 마커 내부 색깔

```python
plt.plot(x,y,color='#a566ff', marker='D', ms='10', mec='b', mew='3', markerfacecolor='y')
```

## 선

- linestyle, ls: 선스타일
<pre>
'-' solid line style
'--' dashed line style
'-.' dash-dot line style
':' dotted line style
</pre>
- linewidth, lw: 선 굵기

```python
plt.plot(x,y,color='#a566ff', marker='D', ms='10', mec='b', mew='3', mfc='y', ls=':', lw='3')
```
